```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# Load the data generated by the C++ backend
df = pd.read_csv('benchmark_data.csv')

# Prepare the data for 3D Surface Plotting
pivot_mr = df.pivot(index='iterations_k', columns='bit_length', values='miller_rabin_time_us')
pivot_aks = df.pivot(index='iterations_k', columns='bit_length', values='aks_time_us')

x_axis = pivot_mr.columns.values  # Bit Lengths
y_axis = pivot_mr.index.values    # Iterations (k)
z_mr = pivot_mr.values            # Miller Rabin Times
z_aks = pivot_aks.values          # AKS / Deterministic Times

# Render the Interactive Plot
fig = go.Figure()

# Add Deterministic (AKS) Surface
fig.add_trace(go.Surface(
    z=z_aks, x=x_axis, y=y_axis,
    colorscale='Reds',
    name='Deterministic (AKS)',
    showscale=False,
    opacity=0.8,
    hovertemplate='<b>Deterministic</b><br>Bit Length: %{x}<br>Iterations: %{y}<br>Time: %{z} µs<extra></extra>'
))

# Add Miller-Rabin Surface
fig.add_trace(go.Surface(
    z=z_mr, x=x_axis, y=y_axis,
    colorscale='Blues',
    name='Miller-Rabin',
    showscale=False,
    opacity=0.9,
    hovertemplate='<b>Miller-Rabin</b><br>Bit Length: %{x}<br>Iterations: %{y}<br>Time: %{z} µs<extra></extra>'
))

# Formatting the layout
fig.update_layout(
    title='Primality Testing: Time Complexity Comparison',
    scene=dict(
        xaxis_title='Bit Length (b)',
        yaxis_title='Iterations (k)',
        zaxis_title='Time Taken (Microseconds)',
        camera=dict(eye=dict(x=1.5, y=-1.5, z=0.5))
    ),
    height=700,
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```
